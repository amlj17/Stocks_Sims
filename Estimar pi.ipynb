```python
import numpy as np
import matplotlib.pyplot as plt
from pandas_datareader import data as pdr
#numero de observaciones
n = 10000
rect = np.random.rand(n,2)
tablero = rect[np.sqrt(rect[:,0]**2+rect[:,1]**2)<1]
estimado = 4*len(tablero)/len(rect)
print('Estimacion de pi = {}'.format(estimado))
plt.figure(figsize=(8,8))
plt.scatter(rect[:,0],rect[:,1], s=0.5, c='blue')
plt.scatter(tablero[:,0], tablero[:,1], s=0.5, c='orange')
```
